```python
# Task 1: Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Task 2: Data frame
df_wine_redwine = pd.read_csv('winequality-red.csv', sep=";")
df_wine_whitewine = pd.read_csv('winequality-white.csv', sep=";")
```

```python
df_wine_redwine['type'] = 'r'
df_wine_whitewine['type'] = 'w'
```

```python
# Combination of two Data frames 
df_wine = pd.concat([df_wine_redwine, df_wine_whitewine], ignore_index=True)
```

```python
df_wine
```

```python
# Task 2: Dataset Statistics
df_wine.describe()
```

```python
# Task 3: Handling missing entries.
for col in df_wine.select_dtypes(include=['object']):
    df_wine[col] = df_wine[col].fillna(df_wine[col].mode()[0])
for col in df_wine.select_dtypes(include=['int64','float64']):
    df_wine[col] = df_wine[col].fillna(df_wine[col].mean())
df_wine
```

```python
# Task 4: Handle mismatched string formats.
for val in df_wine.columns:
    if df_wine[val].dtype == "object":
        df_wine[val] = df_wine[val].str.strip().str.lower()
df_wine
```

```python
# Task 5: Handling outliers.
for _ in df_wine.columns:
    if df_wine[_].dtype in ['float64', 'int64']:
        q1 = df_wine[_].quantile(0.25)
        q3 = df_wine[_].quantile(0.75)
        range__ = q3 - q1
        lower = q1 - 1.5 * range__
        upper = q3 + 1.5 * range__
        mean = df_wine[_].mean()
        df_wine.loc[(df_wine[_] < lower) | (df_wine[_] > upper), col] = mean
df_wine
```

```python
# Task 6: Data Visualization 
Quality = df_wine.groupby("quality").size()
plt.bar(Quality.index,Quality.values)
plt.title("Wine-Quality")
plt.xlabel("quality")
plt.ylabel("count")
plt.xticks(rotation=20)
plt.show()
```

```python
# Task 7: Correlation Identification
for col in df_wine.select_dtypes(include=['object']).columns:
    df_wine[col] = pd.Categorical(df_wine[col]).codes
matrix = df_wine.corr()
correlation = matrix['quality'].sort_values(ascending=False)
print("Correlation value of each feature with quality of wine :\n", correlation)
threshold = 0.04
low = correlation[correlation.abs() < threshold].index
df_wine = df_wine.drop(columns=low)
print(f"\ncolumns with low corelation : {list(low)}")
plt.figure(figsize=(9, 9))
sns.heatmap(df_wine.corr(), cmap='Greens', annot=True, fmt=".2f")
plt.title("Correlation Matrix with Quality", size=10)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
X = df_wine.drop(columns=['density'])
y = df_wine['density']
```

```python
print(y)
```

```python
train_size = int(0.8 * len(X))

X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
normalize = X_train.select_dtypes(include=['float', 'int']).columns

min_train = X_train[normalize].min()
max_train = X_train[normalize].max()

print('Minimum of each column is:\n', min_train)
print('Maximum of each column is:\n', max_train)

for column in normalize:
    X_train[column] = (X_train[column] - min_train[column]) / (max_train[column] - min_train[column])

for column in normalize:
    X_test[column] = (X_test[column] - min_train[column]) / (max_train[column] - min_train[column])

X_train, X_test
```

```python
print("X_train : ",X_train.shape)
print("X_train : ",X_test.shape)
print("y_test : ",y_train.shape)
print("y_test : ",y_test.shape)
```

```python
X_train_b = np.c_[np.ones((X_train.shape[0], 1)), X_train] 
X_test_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
```

```python
weights = np.linalg.inv(X_train_b.T.dot(X_train_b)).dot(X_train_b.T).dot(y_train)
print(weights)
```

```python
y_train_pred = X_train_b.dot(weights)
y_test_pred = X_test_b.dot(weights)
```

```python
mse_train = np.mean((y_train - y_train_pred) ** 2)
mse_test = np.mean((y_test - y_test_pred) ** 2)
# Print MSE in decimal notation
print(f'MSE on training set: {mse_train:.10f}')
print(f'MSE on test set: {mse_test:.10f}')      
```

```python
def mse(weights, filename):
    with open(filename, 'wb') as file:
        pickle.dump(weights, file)
    print(f"Weights saved to {filename}")

mse(weights, 'mse.pkl')
```

```python
#ridge regression
lambda_reg = 0.1

X_train_b = np.c_[np.ones((X_train.shape[0], 1)), X_train]
X_test_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
```

```python
n_features = X_train_b.shape[1]
identity_matrix = np.eye(n_features)
identity_matrix[0, 0] = 0
```

```python
weights_ridge = np.linalg.inv(X_train_b.T.dot(X_train_b) + lambda_reg * identity_matrix).dot(X_train_b.T).dot(y_train)
print(weights_ridge)
```

```python
y_train_pred = X_train_b.dot(weights_ridge)
y_test_pred = X_test_b.dot(weights_ridge)
```

```python
rss=np.sum((y_train - X_train_b.dot(weights_ridge)) **2 )
r=lambda_reg * np.sum(weights_ridge[1:] ** 2)
cost= 0.5 * (rss +r )
print("J(W): ",cost)
```

```python
mse_train_ridge = np.mean((y_train - y_train_pred) ** 2)
mse_test_ridge = np.mean((y_test - y_test_pred) ** 2)

print(f'Ridge Regression MSE on training set: {mse_train_ridge:.10f}') 
print(f'Ridge Regression MSE on test set: {mse_test_ridge:.10f}')     

```

```python
rss=np.sum((y_train - X_train_b.dot(weights_ridge)) **2 )
r=lambda_reg * np.sum(weights_ridge[1:] ** 2)
cost= 0.5 * (rss +r )
print("J(w): ",cost)
```
