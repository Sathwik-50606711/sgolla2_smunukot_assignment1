# Penguins Dataset Part-1 [Preprocessing]

```python
# Task 1: Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Task 2: Data frame
df=pd.read_csv("penguins.csv")
print(df)
```

```python
# Task 2: Dataset Statistics
df.describe()
```

```python
#visualizing the data
df.head()
```

```python
# Task 3: Handling missing entries.
for col in df.select_dtypes(include=['object']):
    df[col] = df[col].fillna(df[col].mode()[0])
for col in df.select_dtypes(include=['int64','float64']):
    df[col] = df[col].fillna(df[col].mean())
df
```

```python
# Task 4: Handle mismatched string formats..
for val in df.columns:
    if df[val].dtype == "object":
        df[val] = df[val].str.strip().str.lower()
df
```

```python
# Task 5: Handling outliers.
for _ in df.columns:
    if df[_].dtype in ['float', 'int']:
        q1 = df[_].quantile(0.25)
        q3 = df[_].quantile(0.75)
        range_ = q3 - q1
        lower = q1 - 1.5 * range_
        upper = q3 + 1.5 * range_
        mean = df[_].mean()
        df.loc[(df[_] < lower) | (df[_] > upper), col] = mean
df
```

```python
# Task 6: Data Visualization 
species = df.groupby("species").size()
plt.bar(species.index,species.values)
plt.title("Count of Penguins")
plt.xlabel("Species")
plt.ylabel("Penguins")
plt.xticks(rotation=20)
plt.show()
```

```python
# Task 6: Data Visualization 
filtered=df[(df["flipper_length_mm"]>180) & (df["flipper_length_mm"]<=210)]
bin=[180,190,200,210]
a=dict(pd.cut(filtered["flipper_length_mm"],bin).value_counts())
ordered_a=dict(sorted(a.items()))

plt.bar([str(flipper_range) for flipper_range in ordered_a.keys()],ordered_a.values())
plt.xlabel("flipper_range")
plt.ylabel("Count")
plt.title("Count of Penguins based on body mass")
plt.xticks(rotation=30)
plt.show()
```

```python
# Task 6: Data Visualization 
numeric_df = df.select_dtypes(include=['float', 'int'])
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(8, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".1f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Task 6: Data Visualization 
sns.kdeplot(x='bill_length_mm', hue='species', data=df)
plt.title('Bill Length - Species')
plt.xlabel('Bill Length')
plt.show()
```

```python
# Task 6: Data Visualization 
species_counts = df['species'].value_counts()
species_counts.plot.pie(startangle=140, colors=sns.color_palette('Set2'), figsize=(6,6))
plt.title('Penguin Species Distribution')
plt.show()
```

```python
# Task 6: Data Visualization 
sns.scatterplot(data=df, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.title('Bill Length - Bill Depth')
plt.xlabel('Bill Length')
plt.ylabel('Bill Depth')
plt.legend(title='Species')
plt.show()
```

```python
# Task 7: Correlation Identification
df['gender'] = pd.Categorical(df['gender']).codes

for col in df.select_dtypes(include=['object']).columns:
    df[col] = pd.Categorical(df[col]).codes
matrix = df.corr()

correlation = matrix['gender'].sort_values(ascending=False)
print("Correlation value of each feature with gender:\n", correlation)

threshold = 0.01

low = correlation[correlation.abs() < threshold].index

df = df.drop(columns=low)

print(f"\ncolumns with low corelation : {list(low)}")


plt.figure(figsize=(9, 9))
sns.heatmap(df.corr(), cmap='Greens', annot=True, fmt=".2f")


plt.title("Correlation Matrix with Gender[Target Variable]", size=10)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Task 7: onehot encoding
df1 = df1.drop(columns=['year', 'island'])

categorical_columns = df1.select_dtypes(include=['object', 'category']).columns.tolist()

df1 = pd.get_dummies(df1, columns=categorical_columns, prefix=categorical_columns)

df1 = df1.astype(int)

df1.head()
```

```python
# Task 9: Normalization

normalize = df1.select_dtypes(include=['float', 'int']).columns
min = df1[normalize].min()
print('minimum of each column is:\n',min)

max = df1[normalize].max()
print('maximum of each column is:\n',max)

for column in normalize:
    df1[column] = (df1[column] - min[column]) / (max[column] - min[column])
df1
```
